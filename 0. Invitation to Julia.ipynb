# Advantages of Julia

Julia is:

- high-level and dynamic

- multi-platform

- free and open-source (MIT)

But, Julia is also
- fast!
- because Just-In-Time (JIT) compiled with sophisticated type system

- and thus solves the infamous **two-language problem**


Julia:

- has a clear type system
- but it is not *necessary* to talk about types

- has *multiple dispatch*: functions specialised on the types of their arguments

- is designed for parallelism and distributed computation

- has a built-in package manager

- Most of the standard library is implemented *in Julia itself* 

- "Users are developers"


- has sophisticated *metaprogramming* (macros, generated functions) for generating code programatically
- allows the creation of domain-specific languages
